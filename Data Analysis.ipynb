# Data Preparation and Model Discovery 

## Missing Data

### Number of Missing Elements per Column 

```python
df.isnull().sum()
```

### Countries having NaN for Agriculture

```python
df[df['Agriculture'].isnull()]['Country']
```

Most of these countries are tiny islands, with the exception of Greenland and the Western Sahara, which are ice-cold or are deserts. These NaN values are filled with zero, since they are extremely small or non-existent. 

```python
df[df['Agriculture'].isnull()] = df[df['Agriculture'].isnull()].fillna(0)
```

### Checking for Climate 

```python
df['Climate'] = df['Climate'].fillna(df.groupby('Region')['Climate'].transform('mean'))
```

### Checking for Literacy 

```python
df[df['Literacy (%)'].isnull()]
```

```python
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

#### Re-checking for Null

```python
df.isnull().sum()
```

#### Dropping the Rest of the countries as there are only few null values 

```python
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

### Final Check 

```python
df.isnull().sum()
```

```python
df = df.dropna()
```
